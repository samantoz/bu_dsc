1) significance
    TF - IDF - Term Frequency - Inverse Document Frequency

    TF = (# of times a term appears in a document / Total # of terms in a document)
    IDF = ln(Total # of documents / # of documents with the term in it) 

    A high weight in tf-idf is reached by a high term frequency (in a given document) and a low frequency in the number of documents that contain the term. As the term appears in more documents, the ratio inside the log approaches 1, bringing the IDF and tf/idf approach 0.

    e.g.,
    assuming each document having 'cat' has it 3 times.
    1) 10000 documents and 1 document contains 'cat'
    (3/100) * ln(10000/1) = 0.2763

    2) 10000 documents and 100 document contains 'cat'
    (3/100) * ln(10000/100) = 0.1381

    3) 10000 documents and all 10000 documents contains 'cat'
    (3/100) * ln(10000/10000) = 0.0

    This helps extracting terms from documents in turn helping to categorize the document.
