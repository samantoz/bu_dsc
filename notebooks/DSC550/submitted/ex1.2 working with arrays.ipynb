1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

3. Determine the minimum; maximum; average; and standard deviation of all entries.

4. Sort the DataFrame: a) sort on column 2; b) sort on row 2.

```python
# Load the library
import numpy as np
import pandas as pd
```

```python
# creating the matrix with a small dataset.
matrix_1 = np.random.randint(1,15,15).reshape(3,5); matrix_1
```

```python
sum_rows = np.sum(matrix_1, axis=0); print('Sum of all rows of the matrix: ', sum_rows)
sum_cols = np.sum(matrix_1, axis=1); print('Sum of all columns of the matrix: ',sum_cols)
sum_all = np.sum(matrix_1);          print('Sum of all elements of the matrix: ',sum_all)
```

```python
# sample size of 20 
# this is doing the full task on a smaller dataset before using a bigger data set
df_test = pd.DataFrame(matrix_1);df_test
```

```python
# sort the dataframe by row 2
df_test.sort_values(by = 2, axis = 1, ascending=True)
```

```python
# sort the dataframe by column 2
df_test.sort_values(by = 2, axis = 0, ascending=True)
```

```python
# This is using the full dataset (1 to 500)
# Setting the seed for regeneration
np.random.seed(1)
# Generate 500 random integers between 1 and 500 (inclusive)
elem = np.random.randint(1,500,500)
elem
```

```python
# Create a matrix of shape [3,2]
matrix = elem.reshape(10,50); matrix
```

```python
# Loading the matrix into a dataframe
df = pd.DataFrame(matrix); df
```

```python
# Calculate the sum of each row; the sum of each column; and the sum of all entries.

sum_rows = np.sum(matrix, axis=0); print('Sum of all rows of the matrix: ', sum_rows)
sum_cols = np.sum(matrix, axis=1); print('Sum of all columns of the matrix: ',sum_cols)
sum_all = np.sum(matrix);          print('Sum of all elements of the matrix: ',sum_all)
```

```python
# Determine the minimum; maximum; average; and standard deviation of all entries.

min_all = np.min(matrix);print('Minimum of all entries: ',min_all)
max_all = np.max(matrix); print('Maximum of all entries: ',max_all)
avg_all = np.mean(matrix); print('Average of all entries: ',avg_all)
sd_all = np.std(matrix); print('Standard dev of all entries: ',sd_all)
```

```python
# Showing the unsorted dataframe once again before sorting by rows and columns
df
```

```python
# Sort dataframe by column 2
df.sort_values(by =2, axis = 0, ascending=True)
```

```python
# sort dataframe by row 2
df.sort_values(by = 2, axis = 1, ascending=True)
```
