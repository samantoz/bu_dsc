```python
# import the relevant libraries

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Using the publicly available gapminder data
# The dataset contains country wise human development data
# Features included are Total Population,Life Expectency and GDP per capita

data_url = 'http://bit.ly/2cLzoxH'
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

```python
# The dataset is of multilple years for all the countries
gapminder.shape
```

```python
# For the convenience of analysis filtering out the data for 2007
gapminder_2007=gapminder[gapminder.year==2007]
```

# Chart 1 - Line Chart
## Rank 1 (Least Effective)

```python
# For initial analysis lets plot the life expectancy against the gdp per capita
# The below chart which is aline chart shows that with increase of GDP percapita the life expectancy increases
# The line chart is very confusing as it shows the data for all countries and they are overlapping

lifeexp = gapminder_2007['lifeExp']
gdppercap = gapminder_2007['gdpPercap']

plt.plot(gdppercap,lifeexp,label='Life Expectency')

plt.xlabel('GDP Per Capita', fontsize = 14)
plt.ylabel('Life Expectency', fontsize = 14)
plt.title('GDP Per Capita vs Life Expectency', fontsize=30)
plt.xticks([0,100,1000, 5000,10000,15000,20000,30000,40000,50000],['0','100','1k', '5k','10k','15K','20k','30k','40k','50k'])
plt.legend()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()
```

# Chart 2 - Pie Chart
## Rank 2

```python
# I am trying to come up with a pie chart
# Now pie chart can only show one feature
# Also if we try to show all countries the pie chart would be too much clumsy
# Hence taking the top 10 countries by population

gapminder_2007_top10 = gapminder_2007.nlargest(10,'pop')
gapminder_2007_top10
# Style of the pie chart
plt.style.use("fivethirtyeight")

plt.pie(gapminder_2007_top10['pop'],
        labels=gapminder_2007_top10['country'],
        shadow=True,
        startangle=90,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'}
    )

plt.title("Top 10 Countries by Population in 2007",fontsize = 30)
plt.tight_layout()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()        
```

# Plot 3 - Scatter Plot
## Rank 3 (Most Effective) 

```python
# Increase the graph size
plt.figure(dpi=150)
#Use seaborn scatterplot for better customization
sns.scatterplot(gapminder_2007['gdpPercap'], gapminder_2007['lifeExp'], hue = gapminder_2007['continent'], size = gapminder_2007['pop'], sizes=(20,400), legend=True)


plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]', fontsize = 14)
plt.ylabel('Life Expectancy [in years]', fontsize = 14)
plt.title('World Development in 2007', fontsize = 30)
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()
```
